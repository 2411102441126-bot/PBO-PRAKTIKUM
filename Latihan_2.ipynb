```python
class Buku:
    def __init__(self, judul, penulis, tahun_terbit):  
        self.judul = judul
        self.penulis = penulis
        self.tahun_terbit = tahun_terbit
        self.status_pinjam = False  # default False, artinya tersedia

    def tampilkan_info(self):
        status = "Dipinjam" if self.status_pinjam else "Tersedia"
        print(f"Judul       : {self.judul}")
        print(f"Penulis     : {self.penulis}")
        print(f"Tahun Terbit: {self.tahun_terbit}")
        print(f"Status      : {status}")
        print("-" * 40)

    def pinjam_buku(self):
        if not self.status_pinjam:
            self.status_pinjam = True
            print(f"Buku '{self.judul}' telah dipinjam.")
        else:
            print(f"Buku '{self.judul}' sedang tidak tersedia.")

    def kembalikan_buku(self):
        if self.status_pinjam:
            self.status_pinjam = False
            print(f"Buku '{self.judul}' telah dikembalikan.")
        else:
            print(f"Buku '{self.judul}' memang belum dipinjam.")

# Bagian utama program
if __name__ == "__main__":   
    # Membuat objek buku
    buku1 = Buku("Rembulan Tenggelam Di Wajahmu", "Tere Liye", 2006)
    buku2 = Buku("Laut Bercerita", "Leila S. Chudori", 2017)

    # Tampilkan informasi awal
    print("=== Informasi Awal Buku ===")
    buku1.tampilkan_info()
    buku2.tampilkan_info()

    # Simulasikan peminjaman 1 buku
    print("\n=== Proses Peminjaman Buku ===")
    buku1.pinjam_buku()

    # Tampilkan kembali informasi setelah dipinjam
    print("\n=== Informasi Setelah Peminjaman ===")
    buku1.tampilkan_info()

    # Simulasikan pengembalian buku
    print("\n=== Proses Pengembalian Buku ===")
    buku1.kembalikan_buku()

    # Tampilkan kembali informasi untuk memastikan status
    print("\n=== Informasi Setelah Dikembalikan ===")
    buku1.tampilkan_info()
```
