```python
# latihan_3.py
# Studi Kasus: Sistem Manajemen Gaji Karyawan (versi modifikasi)

class Karyawan:
    # Constructor
    def __init__(self, nama, id_karyawan, gaji):
        self.__nama = nama
        self.__id_karyawan = id_karyawan
        self.__gaji = gaji

    # Getter
    def get_nama(self):
        return self.__nama

    def get_id_karyawan(self):
        return self.__id_karyawan

    def get_gaji(self):
        return self.__gaji

    # Setter
    def set_nama(self, nama_baru):
        if nama_baru.strip() != "":
            self.__nama = nama_baru
        else:
            print("Error: Nama tidak boleh kosong!")

    def set_gaji(self, gaji_baru):
        if isinstance(gaji_baru, (int, float)) and gaji_baru > 0:
            self.__gaji = gaji_baru
        else:
            print("Error: Gaji harus berupa angka positif!")

    # Method tambahan: hitung bonus
    def hitung_bonus(self, persen_bonus):
        if persen_bonus > 0:
            return self.__gaji + (self.__gaji * persen_bonus / 100)
        else:
            print("Error: Bonus harus lebih besar dari 0")
            return self.__gaji

    # Tampilkan informasi karyawan
    def info_karyawan(self):
        print(f"Nama       : {self.__nama}")
        print(f"ID         : {self.__id_karyawan}")
        print(f"Gaji       : Rp {self.__gaji:,}")


# Bagian Utama Program
if __name__ == "__main__":
    # a. Buat beberapa objek karyawan
    k1 = Karyawan("Budi", "K001", 5000000)
    k2 = Karyawan("Siti", "K002", 4500000)

    print("=== Informasi Awal Karyawan ===")
    k1.info_karyawan()
    k2.info_karyawan()

    # b. Simulasi input salah
    print("\n>> Uji coba gaji negatif")
    k1.set_gaji(-3000000)

    print("\n>> Uji coba nama kosong")
    k2.set_nama("")

    # c. Update data dengan benar
    print("\n>> Update gaji valid")
    k1.set_gaji(6000000)

    print("\n>> Update nama valid")
    k2.set_nama("Siti Aminah")

    # d. Hitung bonus
    print("\n=== Informasi Setelah Bonus ===")
    print(f"{k1.get_nama()} menerima gaji + bonus: Rp {k1.hitung_bonus(10):,}")
    print(f"{k2.get_nama()} menerima gaji + bonus: Rp {k2.hitung_bonus(5):,}")

    # e. Informasi akhir
    print("\n=== Informasi Akhir Karyawan ===")
    k1.info_karyawan()
    k2.info_karyawan()
```
